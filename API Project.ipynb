```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'
params = {"query_text":'fire',
         "year":2016}

response = requests.get(endpoint, params=params)
response.json()
```

```python
fire2016 = pd.DataFrame(response.json())
```

```python
fire2016.sort_values(['query_count'], ascending=False)
```

```python
#2. Make an API request that returns all the times a query was run more than 100 times in a month. 
#How many times did this occur?  
```

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'
params = {"$where":"query_count>=100",
         "year":"2016"}

response2 = requests.get(endpoint, params=params)
response2.json()
```

```python
len(response2.json())
```

```python
url = 'https://data.nashville.gov/resource/fuaa-r5cm.json?$where=query_count>=100'

response3 = requests.get(url)
response3.json()
```

```python
month100 = pd.DataFrame(response3.json())
month100
```

```python
#4. Make an API request that returns the entire Top 500 Monthly Searches dataset.
#Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.

endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'
params = {"query_text":"maps",
          "$limit" : 100000}

response1 = requests.get(endpoint, params = params)
response1.json()
```

```python
mapsdf = pd.DataFrame(response1.json())
mapsdf
```

```python
len(response1.json())
```

```python
mapsdf['query_count'] = mapsdf['query_count'].astype(float)
mapsdf.dtypes
```

```python
plt.figure(figsize=(12,10))
plt.bar('month_name', 'query_count', data = mapsdf)
```
